```python
# Pie Charts: these are good to show the proportion voer the whoel

#creating the dataset
y = np.array([5, 20, 10, 8, 25])
cars = ['Benz' , 'Ford' , 'Honda', 'Tesla', 'Toyota']
spaces = [0, 0.2, .1, .25, 0.1]
colrs = ['b', 'c', 'm', 'y', 'k']
# plotting the dataset:
plt.pie(y, labels = cars, explode = spaces, colors = colrs)
plt.legend();
```

```python
# Let us create a Bivariate scatter plot using Age and Fare variables

plt.scatter(titanic['Age'], titanic['Fare'], s=.5, color='r');
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Fare to Age Distribution')
plt.xticks(range(0, 105, 5));
```

### Observations

1. **Relationship Between Variables**: A scatter plot helps identify the relationship between two variables. If the points form a clear upward or downward trend, it suggests a positive or negative correlation, respectively.

2. **Clusters or Grouping**: The plot can show if the data points form distinct clusters or groups, which might indicate the presence of different subgroups or patterns within the data.

3. **Outliers**: Scatter plots can highlight outliersdata points that don't follow the general pattern of the rest of the data. Outliers appear as points that are far away from the main cluster of points.

```python
# Let's compare Sex and Age variables using a boxplot
# sns calls seaborn
# Needs x , y and name of dataset assigned
sns.boxplot(x = 'Sex', y = 'Age', data=titanic, color='cyan');
```

# Observation


1. **Median (Central Tendency)**: The line inside the box represents the median, which is the middle value of the dataset. It divides the dataset into two equal halves.

2. **Interquartile Range (IQR)**: The box itself represents the interquartile range, which is the range between the first quartile (25th percentile) and the third quartile (75th percentile). This shows where the central 50% of the data falls.

3. **Whiskers**: The lines extending from the box (whiskers) represent the range of the data. The whiskers typically extend to 1.5 times the IQR above and below the box. Data points beyond this range are considered potential outliers.

4. **Outliers**: Any data points that fall outside of the whiskers (usually beyond 1.5 times the IQR) are considered outliers. These are shown as individual points or dots outside the whiskers.

5. **Skewness**: The box plot can help identify skewness in the data. If the median line is closer to the top or bottom of the box, or if the whiskers are uneven, it indicates that the data is skewed (either positively or negatively).

# Heatmap Graph

**Heatmaps:**

1. Heatmaps are a seaborn library plot that display correlation.
2. What is correlation? It is the relationship between two variables.
3. Correlation can be positive or negative.
4. If the variables move together in either direction (up or down), we call this a positive correlation.
5. If the variables move in opposite directions of one another then this would be called a negative correlation.
6. A heat map is designed for numeric values only.
7. The range of the correlation in the heatmap goes from -1 to +1

```python
# Let's create a new dataset and name it num_df that contains only numberic variables of titanic

num_df = titanic[['Age', 'Fare', 'Survived', 'Pclass', 'SibSp', 'Parch']]
num_df
```

```python
num_df.corr()
```

```python
# correlation table
num_df.corr()
```

**Observations**
1, **Introduction to Heatmaps**: Heatmaps are used to visualize correlation between variables. I learned that the range of correlation is from -1 to +1.

2. **Creating a Numeric Dataset**: I created a new dataset (`num_df`) containing only numeric variables like `Age`, `Fare`, `Survived`, etc., to analyze correlations.

3. **Correlation Calculation**: The `.corr()` function was used to calculate the correlation between different numeric variables, like `Age` and `Fare`.

4. **Creating a Heatmap**: I generated a heatmap to visualize the correlation matrix. This helps identify which variables are strongly correlated.

5. **Customization of Heatmap**: I used `sns.heatmap()` with options like `annot=True` to annotate the heatmap and `cmap='RdYlGn'` to set the color map, making the correlations visually more interpretable.

```python
# Let's create the heatmap for this dataset
plt.figure(figsize = (12,8)) #settings the size of the heat map
sns.heatmap(num_df.corr(), annot=True, cmap= 'RdYlGn');
```
